```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
from scipy.stats import f_oneway

datav1 = df['V1'].to_numpy()
datav2 = df['V2'].to_numpy()

datav1_c1 = datav1[all_predictions == 0]
datav1_c2 = datav1[all_predictions == 1]
datav1_c3 = datav1[all_predictions == 2]

datav2_c1 = datav2[all_predictions == 0]
datav2_c2 = datav2[all_predictions == 1]
datav2_c3 = datav2[all_predictions == 2]

print(f_oneway(datav1_c1, datav1_c2, datav1_c3))
print(f_oneway(datav2_c1, datav2_c2, datav2_c3))
```

Переменные по которым были обнаруженны значимые различия между выделенными кластерами: V2 (pvalue < 0.05)

```python
pca = PCA(n_components=2)

def get_pc(df):
    return df.assign(PC1=pca.fit_transform(df)[:, 0],
           PC2=pca.fit_transform(df)[:, 1])

df2 = get_pc(df)
df2
```

```python
def get_pca2(df, threshold=0.9):
    pca = PCA()
    fit = pca.fit_transform(df)
    n_components = sum(np.cumsum(pca.explained_variance_ratio_) < threshold) + 1
    summ = 0
    k = 0
    for a in pca.explained_variance_ratio_:
        print(float(a))
        if summ < 0.9:
            summ+=float(a)
            k+=1
    print()
    print(k,"- Минимальное количество компонент для обьяснения 90% изменчивости")
    print()
    return pd.concat([df, pd.DataFrame(fit[:, :n_components])], axis=1)

print(get_pca2(df2))

```

```python
def is_multicol(df):
  corr = df.corr()
  triu= pd.DataFrame(np.triu(corr,k=0))
  per = []
  for i in range(0, len(corr)):
    for j in range(0, len(corr)):
      if (triu.iloc[i,j]==1) & (i != j):
        per.append(j+1)
        per.append(i+1)
  if per == []:
    print("There is no collinearity in the data")
  else:
    i=0
    while i < len(per):
      print('V'+str(per[i])+' '+'V'+str(per[i+1]),end=' ')
      i += 2
is_multicol(df)
```
